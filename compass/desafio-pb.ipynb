### Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Lendo arquivo CSV

```python
df = pd.read_csv("./data/googleplaystore.csv")
```

#### Removendo items duplicados

```python
df = df.drop_duplicates(subset=['App'])
```

### Gráfico de barras com os cinco apps com o maior número de instalação

#### Mudando o tipo da coluna Install de object para float64

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', ''))
df['Installs'] = df['Installs'].apply(lambda x: x.strip('+'))
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
```

#### Criando tabela

```python
top_installed_apps = df.nlargest(5,'Installs')

plt.bar(top_installed_apps['App'], top_installed_apps['Installs'])

plt.title("Aplicações mais Instaladas do Google PlayStore")
plt.xlabel('Aplicativos')
plt.ylabel('Número de instalações (arredondados por bilhão)')
plt.xticks(rotation=45)

plt.show()
```

### Gráfico de categorias com a frequência em que elas aparecem

```python
category_counts = df['Category'].value_counts()

labels = category_counts.index.tolist()
sizes = category_counts.values.tolist()

plt.figure(figsize=(17, 17))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', pctdistance=1.05, labeldistance=.6)
plt.title('Frequência das Categorias')
plt.show()
```
